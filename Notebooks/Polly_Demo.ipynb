```python
import boto3
import IPython
from pprint import pprint

polly = boto3.client('polly', region_name='eu-west-1')
```

```python
response = polly.synthesize_speech(
    Text="Guten Tag meine sehr verehrten Damen und Herren, mein Name ist Hans!",
    TextType="text",
    OutputFormat="mp3",                                           
    VoiceId="Hans")

pprint (response)
     
outfile = "pollyresponse.mp3"
data = response['AudioStream'].read()

with open(outfile,'wb') as f:
     f.write(data)
IPython.display.Audio(outfile) 
```
